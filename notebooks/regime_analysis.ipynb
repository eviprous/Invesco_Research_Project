```python
cw_ret = (cw * cw_weights).sum(axis=1)
```

```python
cw_cum = (1 + cw_ret).cumprod() - 1
```

```python
window = 252  # 90 or 252
var_m = mkt.rolling(window).var()
beta = cw.rolling(window).cov(mkt).div(var_m, axis=0)
```

```python
beta_m = beta.resample("M").last()
```

```python
from scipy.optimize import minimize

def equal_abs_beta_contribution_weights(beta_row):
    #filter out nans and zero betas
    valid_betas = beta_row.dropna()
    valid_betas = valid_betas[valid_betas != 0]

    if valid_betas.empty:
        return pd.Series(dtype=float)

    betas = valid_betas.values
    n = len(betas)
    idx = valid_betas.index

    #objective function to minimize with target of 1/n
    def objective(w):
        contrib = w * betas
        return np.sum((contrib - 1/n)**2)

    #constraints: sum of weights = 1
    constraints = [
        {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},
    ]

    #long only weights to mimic equal weight
    bounds = [(0, 1) for _ in range(n)]

    #start with inverted betas
    x0 = 1 / np.abs(betas)
    x0 = x0 / x0.sum()

    result = minimize(objective, x0, bounds=bounds, constraints=constraints)

    if not result.success:
        return pd.Series(dtype=float)

    weights = pd.Series(result.x, index=idx)

    full_weights = pd.Series(0.0, index=beta_row.index)
    full_weights.update(weights)

    return full_weights
```

```python
import statsmodels.api as sm

idx = cw_ret.index.intersection(ff.index)
rf = ff.loc[idx, "RF"]
cw_excess = cw_ret.loc[idx] - rf

has_ebc = ("ebc_ret" in globals() and ebc_ret is not None)
if has_ebc:
    idx = idx.intersection(ebc_ret.index)
    rf = ff.loc[idx, "RF"]
    cw_excess = cw_ret.loc[idx] - rf
    ebc_excess = ebc_ret.loc[idx] - rf

# CAPM
X_capm = sm.add_constant(ff.loc[idx, ["MKT_RF"]])
capm_cw = sm.OLS(cw_excess, X_capm).fit(cov_type="HAC", cov_kwds={"maxlags": 5})

capm_params = pd.DataFrame({"CW": capm_cw.params})
if has_ebc:
    capm_ebc = sm.OLS(ebc_excess, X_capm).fit(cov_type="HAC", cov_kwds={"maxlags": 5})
    capm_params["EBC"] = capm_ebc.params

plt.figure(figsize=(6,4))
capm_params.drop(index="const").plot(kind="bar")
plt.axhline(0)
plt.title("CAPM Beta (HAC)")
plt.grid(True, axis="y")
plt.tight_layout()
plt.show()

# FF5+MOM
factors = ["MKT_RF","SMB","HML","RMW","CMA","MOM"]
X_ff = sm.add_constant(ff.loc[idx, factors])

ff_cw = sm.OLS(cw_excess, X_ff).fit(cov_type="HAC", cov_kwds={"maxlags": 5})
coef = pd.DataFrame({"CW": ff_cw.params}).drop(index="const")

if has_ebc:
    ff_ebc = sm.OLS(ebc_excess, X_ff).fit(cov_type="HAC", cov_kwds={"maxlags": 5})
    coef["EBC"] = ff_ebc.params

coef = coef.reindex(factors)

plt.figure(figsize=(10,4))
coef.plot(kind="bar")
plt.axhline(0)
plt.title("FF5 + MOM Loadings (HAC)")
plt.grid(True, axis="y")
plt.tight_layout()
plt.show()

# Alpha annual
alpha_ann = pd.Series({"CW": ff_cw.params["const"] * 252})
if has_ebc:
    alpha_ann["EBC"] = ff_ebc.params["const"] * 252

plt.figure(figsize=(6,4))
plt.bar(alpha_ann.index, alpha_ann.values)
plt.axhline(0)
plt.title("Annualized Alpha (FF5+MOM, const*252)")
plt.grid(True, axis="y")
plt.tight_layout()
plt.show()
```
